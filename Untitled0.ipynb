```python
"""
PROJET DATA SCIENCE - MODULE COLLECTE (WEB SCRAPING)
Rôle : Data Engineer
Description : Script complet avec gestion d'encodage pour compatibilité Excel.
"""

import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import os
import re

# URL de la catégorie
url = "http://books.toscrape.com/catalogue/category/books/travel_2/index.html"
headers = {"User-Agent": "Mozilla/5.0"}

def extraire_tous_les_produits():
    try:
        reponse = requests.get(url, headers=headers, timeout=10)
        reponse.raise_for_status()

        soup = BeautifulSoup(reponse.text, 'html.parser')
        articles = soup.find_all('article', {'class': 'product_pod'})

        liste_donnees = []

        for article in articles:
            # Extraction du titre
            titre = article.h3.a['title']

            # Extraction du prix
            prix_brut = article.find('p', {'class': 'price_color'}).text

            # Nettoyage robuste
            prix_net = re.sub(r'[^\d.]', '', prix_brut)
            prix_propre = float(prix_net)

            liste_donnees.append({
                'Date': datetime.now().strftime("%Y-%m-%d"),
                'Produit': titre,
                'Prix': prix_propre
            })

        df = pd.DataFrame(liste_donnees)

        # 4. Enregistrement en CSV avec l'encodage correct pour Excel
        nom_fichier = 'liste_produits_complet.csv'
        fichier_existe = os.path.isfile(nom_fichier)

        # 'utf-8-sig' est la clé pour que Excel affiche bien les caractères spéciaux
        df.to_csv(nom_fichier, mode='a', index=False, header=not fichier_existe, encoding='utf-8-sig')

        print(f"Succès ! {len(df)} produits extraits et enregistrés dans {nom_fichier}.")

    except Exception as e:
        print(f"Une erreur est survenue : {e}")

if __name__ == "__main__":
    extraire_tous_les_produits()
```
